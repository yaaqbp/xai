```python
def flatten_dict(variable_groups):
    new_dict  = {}
    for key in variable_groups.keys():
        for value in variable_groups[key]:
            new_dict[value] = key
    return new_dict
```

```python
def group_plot(shap_prediction, variable_groups = variable_groups_eng):
    baseline = shap_prediction.intercept
    df = shap_prediction.result
    df['group'] = df['variable_name']
    df.replace({'group':flatten_dict(variable_groups)}, inplace = True)
    sum_df = df.groupby('group').sum()[['contribution']]
    sum_df['abs_contribution'] = abs(sum_df['contribution'])
    sum_df.sort_values(by = 'abs_contribution', ascending = True, inplace = True)
    sum_df = sum_df[sum_df.index.isin(list(variable_groups.keys()))]
    #return plt.barh(y = sum_df.index, width = sum_df.contribution)
    sum_df['sign'] = np.where(sum_df.contribution >0, 1, -1).astype('str')
    
    vcolors = ["#371ea3", "#8bdcbe", "#f05a71"]
    i = len(sum_df.index) 
    sum_df.contribution = round(sum_df.contribution, 2)
    fig = px.bar(title = 'Shapley Values',
             labels = {'group':''}
    )
    
    fig.add_shape(
            type='line',
            x0=baseline,
            x1=baseline, 
            y0 = -1,
            y1 = 13,
            line={'color': "#371ea3", 'width': 1.5, 'dash': 'dot'},
            xref="x"
    )
    
    fig.add_bar(
        orientation = 'h',
        x = sum_df.contribution, 
        y = sum_df.index,
        base = baseline,
        text = sum_df.contribution,
        textposition = 'outside',
        marker_color=[vcolors[int(c)] for c in sum_df['sign'].tolist()],
        hoverlabel={'bgcolor': 'rgba(0,0,0,0.8)'},
)
    fig.update_layout(showlegend=False,
                      margin={'t': 78, 'b': 71, 'r': 30}
                      #paper_bgcolor='rgba(0,0,0,0)',
                    #plot_bgcolor='rgba(0,0,0,0)'
                 )

    fig.update_yaxes({'griddash':'solid',
                        'type': 'category', 
                      'autorange': 'reversed', 
                      'gridwidth': 2, 
                      'automargin': True,
                      'ticks': 'outside', 
                      'tickcolor': 'white', 
                      'ticklen': 10, 
                      'fixedrange': True},
                        row = 1 + i, col=1)
    fig.update_xaxes({'type': 'linear', 'gridwidth': 2, 'zeroline': False, 'automargin': True,
                              'ticks': "outside", 'tickcolor': 'white', 'ticklen': 3, 'fixedrange': True},
                             row=i + 1, col=1)

    
    cum = sum_df.contribution.values + baseline
    min_max_margin = cum.ptp() * 0.15
    temp_min_max = [np.Inf, -np.Inf]
    temp_min_max[0] = np.min([temp_min_max[0], cum.min() - min_max_margin])
    temp_min_max[1] = np.max([temp_min_max[1], cum.max() + min_max_margin])
    fig.update_xaxes({'range': temp_min_max},showline = True)
    return fig 
```
